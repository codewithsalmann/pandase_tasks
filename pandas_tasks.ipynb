```python
# Drop rows with missing values and display the new shape
df = df.dropna()
print('Shape after dropping missing values:', df.shape)
df.head()
```

```python
# Convert 'price' column to numeric and display data types
df['price'] = pd.to_numeric(df['price'], errors='coerce')
print(df.dtypes)
df.head()
```

```python
# Remove duplicate records and display the new shape
df = df.drop_duplicates()
print('Shape after removing duplicates:', df.shape)
df.head()
```

```python
# Find the most common car brands in the dataset
if 'brand' in df.columns:
    top_brands = df['brand'].value_counts()
    print('Top car brands:')
    print(top_brands)
else:
    print('No column named "brand" found in the dataset.')
```

```python
# Visualize the top car brands with a bar chart
import matplotlib.pyplot as plt
if 'brand' in df.columns:
    top_brands = df['brand'].value_counts()
    plt.figure(figsize=(10,6))
    top_brands.plot(kind='bar')
    plt.title('Number of Listings per Car Brand')
    plt.xlabel('Car Brand')
    plt.ylabel('Number of Listings')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print('No column named "brand" found in the dataset.')
```

```python
# Plot a histogram to visualize the distribution of car prices
import matplotlib.pyplot as plt
if 'price' in df.columns:
    plt.figure(figsize=(10,6))
    plt.hist(df['price'].dropna(), bins=30, color='skyblue', edgecolor='black')
    plt.title('Distribution of Car Prices')
    plt.xlabel('Price')
    plt.ylabel('Number of Listings')
    plt.tight_layout()
    plt.show()
else:
    print('No column named "price" found in the dataset.')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

#  Step 1: Define the DataFrame (sample data)
data = {
    'year': [2005, 2008, 2010, 2012, 2015, 2018, 2020, 2022],
    'price': [5000, 7000, 9000, 11000, 13000, 16000, 20000, 22000]
}
df = pd.DataFrame(data)

#  Step 2: Check for required columns and clean data
if 'price' in df.columns and 'year' in df.columns:
    df_clean = df[['year', 'price']].dropna()

    #  Step 3: Group by year and calculate mean price
    year_price = df_clean.groupby('year')['price'].mean().reset_index()

    #  Step 4: Plot the trend
    plt.figure(figsize=(10, 6))
    plt.plot(year_price['year'], year_price['price'], marker='o', linestyle='-')
    plt.title('Average Car Price by Year')
    plt.xlabel('Year')
    plt.ylabel('Average Price')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    #  Step 5: Calculate correlation coefficient
    corr, p_value = pearsonr(df_clean['year'], df_clean['price'])

    #  Step 6: Interpret the result
    if corr > 0.2:
        print(f' Yes, newer cars generally cost more. (Correlation coefficient: {corr:.2f})')
    elif corr < -0.2:
        print(f' No, newer cars generally cost less. (Correlation coefficient: {corr:.2f})')
    else:
        print(f' There is no strong trend between year and price. (Correlation coefficient: {corr:.2f})')
else:
    print(' Required columns "price" and/or "year" not found in the dataset.')

```
